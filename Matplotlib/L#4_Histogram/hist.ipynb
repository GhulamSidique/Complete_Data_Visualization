### A histogram is a graphical representation of a grouped frequency distribution with continuous classes. It is an area diagram and can be defined as a set of rectangles with bases along with the intervals between class boundaries and with areas proportional to frequencies in the corresponding classes.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating a random array
n=np.random.randint(10, 50, (40)) # 40 values between 10 to 50
n
```

```python
x=[31, 19, 31, 23, 48, 38, 22, 13, 27, 46, 40, 17, 43, 16, 29, 40, 36,20, 28, 39, 11, 38, 27, 47, 33, 35, 45, 25, 31, 24, 27, 38, 13, 40, 29, 39, 32, 12, 45, 28]

# creating a simple hist plot
plt.hist(x, color='y', edgecolor='black')

plt.title("Simple histogram")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
# create bins to keep data between the ranges
bins=[10, 20, 30, 40]

plt.hist(x, color='y', edgecolor='black', bins=bins)

plt.title("Simple histogram")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
# we can also adjust the auto bins and keep the values between the ranges
plt.hist(x, 'auto', (1, 200), edgecolor='black')

plt.title("Simple histogram")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
# to make ethe graph start from 10 instead of 0 make use of the bottom = 10 parameter
plt.hist(x, color='y', edgecolor='black', bottom=10)

plt.title("Simple histogram")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
# create horizontal histogram using orientation = 'horizontal'
plt.hist(x, color='y', edgecolor='black', orientation='horizontal')

plt.title("Simple histogram")
plt.xlabel("Y values")
plt.ylabel("X values")

plt.show()
```

```python
# separate the ranges using rwidth parameter
plt.hist(x, color='y', edgecolor='black', rwidth=0.8)

plt.title("Simple histogram")
plt.xlabel("Y values")
plt.ylabel("X values")

plt.show()
```

```python
# display histo in log format
# also introducing legend here
plt.hist(x, color='y', edgecolor='black', log=True, label='Histogram')

plt.title("Simple histogram")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.legend()
plt.show()
```

```python
# to cut the histogram at a specific range make use of plt.axvline(parameters)
# we will also show grid here using plt.grid()
plt.hist(x, color='y', edgecolor='black', label='Histogram')

plt.title("Simple histogram")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.axvline(30, color='r', label='separation line') # this will draw a line at 30 range values

plt.grid()
plt.legend()
plt.show()
```
