### save plot as a figure in your local machine using python code

```python
import matplotlib.pyplot as plt
```

```python
# make some sample arrays
x=[1,3,4,5,6,7,8]
y=[2,5,7,8,9,5,6]

# plot a simple line graph
plt.plot(x,y, color='r')
plt.title("simple line graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# now save the image in the folder
# the image will be save in the current working directory where the current code is running

plt.savefig("line graph", dpi=2000, facecolor='b') # the image name will be the line graph
# the dpi = any big number--> increases the quality of the graph
# the facecolor='b'--> change the fig's boundary color


# to save the image in pdf format
plt.savefig("line graph.pdf", dpi=2000)
# therefore, to convert file type just write the type of the file like .png, .jpg etc

plt.show()
```

```python
# to draw a transparanet graph make use of the transparent=True parameter as follows

plt.plot(x,y, color='r')
plt.title("simple line graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# now save the image in the folder
# the image will be save in the current working directory where the current code is running

plt.savefig("transparent line graph", dpi=2000, facecolor='y', transparent=True) 

plt.show()
```
