### A stem plot is a graphical way of displaying quantitative data1. It consists of a stem, which is the highest place value digit of the data, and leaves, which are the remaining digits2. A stem plot shows the shape of a distribution and the individual values of the data1. It is similar to a histogram, but uses actual numbers instead of blocks2.

```python
import matplotlib.pyplot as plt
```

```python
# make two sample arrays
x=[1,3,5,7,9]
y=[2,4,6,8,10]

plt.stem(x, y)
plt.title("Simple stem plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# to change the shape of the line make use of the linefmt parameter
# to chnage the marker along with its color make use of markerfmt parameter
# to start the stem graph from any number make use of the bottom parameter
# to chnage base line color make use of the basefmt parameter
# bydefault the graph is ploted vertically, to make it horizontally make use of the orientation parameter

plt.stem(x, y, linefmt=":", markerfmt="r*", bottom=5, basefmt='y', orientation='horizontal') # 'r' is for red color
# graph will start from 5 on y axis

plt.title("Simple stem plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.legend()
plt.show()
```
