### here we will be working with xticks, yticks, xlim and ylim
### these functions are used to change the values graph accordingly

```python
import matplotlib.pyplot as plt
```

```python
# make some sample arrays
x=[1,3,5,6]
y=[3,1,2,5]

# line graph
plt.plot(x, y, color='y')
plt.title("smaple graph")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
# as in the above graph the x and y axis are labeled as  numbers and floats respectively
# to change these ticks we have 

# line graph
plt.plot(x, y, color='y')
plt.title("smaple graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# change the x and y tikcs
plt.xticks(x, labels=['py', 'c', 'java', 'cpp'])
plt.yticks(y)

plt.show()
```

```python
# now to add limit of values on the x and y axis
# line graph
plt.plot(x, y, color='y')
plt.title("smaple graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# x axis limit 
plt.xlim(0, 10)
plt.ylim(1, 9)

plt.show()
```

```python
# limit can also be added using axis function as follows
# line graph
plt.plot(x, y, color='y')
plt.title("smaple graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# add axis limit
plt.axis([0,10, 1,9]) # here 0,10 is for x axis and 1,9 is for y axis

plt.show()
```
