### A Step plot is a Line chart that uses vertical and horizontal lines to connect the data points in a series forming a step-like progression1. The vertical parts of a Step chart denote changes in the data and their magnitude1. The step() function is used to design the plot such that it has a horizontal baseline to which the data points will be connected by vertical lines2. This kind of plot is used to analyze at which points the change in Y-axis value has occurred exactly with respect to X-axis

### mostly used in communication and works as signals

```python
import matplotlib.pyplot as plt
```

```python
# maake sample arrays
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
# make changes in the values of the y
y1=[2,4,1,3,2]
x=[1,2,3,4,5]

# to change the step colors make use of the color='color name' parameter
# to change the marker make use of the marker='shape' parameter
# to change the marker size make use of the ms=value parameter
# to change the markere color make use of the mfc='name' parameter
# to add lable make use of the label='name' and also add legend() as well

plt.step(x,y1, color='y', marker='o', ms=10, mfc="g", label='python')
plt.title("sample step plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.legend()
plt.show()
```
