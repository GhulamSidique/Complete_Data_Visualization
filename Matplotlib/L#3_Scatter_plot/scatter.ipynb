### A scatter plot is a graph that presents the relationship between two variables in a data-set12. It represents data points on a two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis1.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating two random arrays
x=np.random.randint(low= 0, high = 10, size=(10,))
y=np.random.randint(low= 10, high = 20, size=(10,))

# plot a simple scatter plot using above random arrays
plt.scatter(x, y, color='r')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# we can use aplha, edgecolor, linewidth parameter as we had used it in the bar plot folder.
# change the dot shape using marker='any shape like +, -, *, etc'
plt.scatter(x, y, color='r', marker='*')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# now to introduce a color bar by the side of the scatter graph by introduncing different colors ford the dots as follows
plt.scatter(x, y, color='r')

plt.colorbar().set_label("color bar")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
