### A pie chart is a circular statistical graphic that illustrates numerical proportion12345. It is divided into slices, with each slice representing a portion of the whole. The size of each slice corresponds to the quantity it represents.

```python
import matplotlib.pyplot as plt
```

```python
# creating two sample arrays
x=[10, 20, 30, 40]
y=['c', 'cpp', 'java', 'python']

# plot a simple graph
plt.pie(x, labels=y) # here labels will be drwan over each slice
plt.title("Pie plot")
plt.show()
```

```python
# to separate a slice from the pie chart just make use of the explode parameter
# to keep our own colors we will pass colors parameter
# to show the percentage value of each slice we will make use of the autopct parameter
# to make shadow down our pie chart use shadow=True

ex=[0.1, 0, 0, 0] # here 0.2 will separate the c from the pie
c=['r', 'b', 'g', 'y']

plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.1f%%", shadow=True)
plt.title("Pie plot")
plt.show()
```

```python
# to change the size of the chart make use of the radius parameter
# to provide distance between the labels and the chart make use of the labeldistance parameter
# bydefault the chart starts from 0 degree angle and to change its starting angle make use of the startangle parameter
# to chnage the size of the fonts make use of the textprops parameter and pass values in the dictionary 
# to shnage the color of the edge make use of the wedgeprops parameter and pass values in a dictionary
# to change or rotate the labels make use of the rotatelabels parameter and make it true or false
# to chnage the location of the lable make use of the loc parameter inside legend()


plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=True, radius=1.5, labeldistance=1.2, startangle=90, textprops={'fontsize': 15}, wedgeprops={"edgecolor": "b"}, rotatelabels=True)
plt.title("Pie plot")

plt.legend(loc=2)
plt.show()
```

```python
# to make overlaping multiple charts 

x=[10, 20, 30, 40]
y=['c', 'cpp', 'java', 'python']
x1=[15, 25, 35, 45]

# plot a simple graph
plt.pie(x, labels=y, radius=1.5)
plt.pie(x1, labels=y, radius=0.5)

plt.title("Pie plot")
plt.show()
```

```python
# clear the center of the pie chart

x=[10, 20, 30, 40]
y=['c', 'cpp', 'java', 'python']

# plot a simple graph
plt.pie(x, labels=y, radius=1.5)
plt.pie([1],colors='w', radius=1)

plt.show()
```
