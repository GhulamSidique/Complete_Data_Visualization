### in this notebook we ill be working with text inside and over the graph

```python
import matplotlib.pyplot as plt 
```

```python
# make arrays
x=[1,3,5,6]
y=[3,1,2,5]

# line graph
plt.plot(x, y, color='y')
plt.title("smaple graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# make a box at any position on the graph as follows
plt.text(3,1, "java", bbox={'facecolor': 'g'})
# here 3,1 shows the x and y axis and position where the box will be created, the name of that box will be java and the color of the box will be green as defined above

plt.show()
```

```python
# to indicate the position with an arrow make use of the arrowprops parameter and .annotate functionas follows
# line graph
plt.plot(x, y, color='y')
plt.title("smaple graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# show position with an arrow
plt.annotate("python", xy=(3, 1.5), xytext=(4,3), arrowprops=dict(facecolor="black"))
# here, xy position is where the python lies and the name python will be displayed on the position 4, 3 on the graph, the color of the arrow will be black

plt.show()
```
