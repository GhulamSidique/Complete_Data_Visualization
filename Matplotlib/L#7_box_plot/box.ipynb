### It is a type of chart that depicts a group of numerical data through their quartiles. It is a simple way to visualize the shape of our data. It makes comparing characteristics of data between categories very easy.

### A box plot gives a five-number summary of a set of data which is-

### Minimum – It is the minimum value in the dataset excluding the outliers.
### First Quartile (Q1) – 25% of the data lies below the First (lower) Quartile.
### Median (Q2) – It is the mid-point of the dataset. Half of the values lie below it and half above.
### Third Quartile (Q3) – 75% of the data lies below the Third (Upper) Quartile.
### Maximum – It is the maximum value in the dataset excluding the outliers.

```python
import matplotlib.pyplot as plt
```

```python
# make a sample array
x=[10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python
# to make a horizontal plot make use of the vert = False parameter
# to change the width of the box  make use of the widths = any number parameter
# to add label at the bottom of the box make use of the label = ["any name"] parameter
# to fill color inside the box make use of the patch_artist=True parameter
# to display the mean position make use of the showmeans = True parameter
# to chnage the boundary color make use of the boxprops=dict(color='any name') parameter
# to chnage the color of the upper and lower caps make use of the capprops=dict(color='color name') parameter
# to chnage the color of the line between whisker and cap make use of the wiskerprops=dict(color='color name') parameter


plt.boxplot(x, vert= False, widths=0.5, labels=['python'], patch_artist=True, showmeans=True, boxprops=dict(color='r'), capprops=dict(color='b'), whiskerprops=dict(color='y'))

plt.show() 
```

```python
# working with an outlier in the box plot
y=[10,20,30,40,50,60,70, 120] # 150 is an outlier here

# to change the shape of the mean point make use of the sym="*,-,+" parameter

plt.boxplot(y, sym='r*')
plt.show()
```

```python
# to fill the gap between above outlier and graph just make use of the whis=any small number unless it fills the gap, parameter

plt.boxplot(y, whis=3.5) # 3.5 is the value that fills the gap between outlier and the graph
plt.show()
```

```python
# making multiple graphs
y=[10,20,30,40,50,60,70, 120]
y1=[15, 25, 35, 45, 55, 65, 75, 85]

to_plot=[y, y1]

plt.boxplot(to_plot, labels=['python', 'java'])
plt.show()
```
