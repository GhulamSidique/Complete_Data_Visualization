### Subplots are used to create multiple non overlapping graphs

```python
import matplotlib.pyplot as plt
```

```python
# make some sample arrays
x=[1,3,5,7,9]
y= [2,4,6,8,10]
z=['a', 'b', 'c', 'd', 'e']

# plot a line graph
plt.subplot(2,2,1)# here first 2 means row number and second shows the column number and finally the 1 shows the position
plt.plot(x,y, color='r')

# ploting a simple pie graph
plt.subplot(2,2,2)
plt.pie([1], colors='g')

# plot a bar chart
plt.subplot(2,2,3)
plt.bar(x, z)


plt.show()
```
