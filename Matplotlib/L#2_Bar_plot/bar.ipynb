### Bar graphs are the pictorial representation of data (generally grouped), in the form of vertical or horizontal rectangular bars, where the length of bars are proportional to the measure of data.

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# make an array of programming languages to show their popularity level
# array of programming language names
langs=['python', 'java', 'c++', 'c', 'js']
# array of popularity
popu=[90, 70, 88, 77, 53]

#plot the simple bar graph
plt.bar(langs, popu, color='y')

plt.title("Programming languages popularity graph")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.show()
```

```python
# we can define width which represents the distnace between the bars
# we can define multiple colors in an array and make use of it in the graph as follows
# we can introduce align='edge' parameter which is used to draw an edge at the beginning of every xlabel. wherease, align='center' will draw an edge at the center of the xlable
# to change the boundary color of every bar just introduce edgecolor='any color name'
# to increase the line/boundary width just ,make use of linewidth=any value
# we can change the boundary style as well by writing linestyle='any shape like : etc'
# to make the graph's color light, write alpha=any value in decimal

colors=['r', 'y', 'g', 'm', 'b']

# now again ploting graph
plt.bar(langs, popu, width=0.4 ,color=colors, align='edge', edgecolor='black', linewidth=5, linestyle=':', alpha=0.4)

plt.title("Programming languages popularity graph")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.show()
```

### legend: it is used to introduce a legend inside the graph as follows

```python
# to make use of the legend we will write label='any name' and over the plt.show() wirte plt.legend() 
colors=['r', 'y', 'g', 'm', 'b']

plt.bar(langs, popu, width=0.4 ,color=colors, edgecolor='black', label='Bar graph')

plt.title("Programming languages popularity graph")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.legend()
plt.show()
```

### creating multiple overlapping bar graphs

```python
# make an extra array of popularity
pop2=[20,30,40,50,60]

plt.bar(langs, popu, color='r', label='Bar graph1')
plt.bar(langs, pop2, color='y', label='Bar graph2')

plt.title("Programming languages popularity graph")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.legend()
plt.show()
```

### creating non overlapping multiple graphs 

```python
# for this we need to create an np array consists of indices equal to the length of the langs array
a=np.arange(len(langs))

# we will also make use of the width
width=0.2

# now make a gap between two graphs by adding indices values with the width
a=[j+width for j in a] 


pop2=[20,30,40,50,60]

plt.bar(langs, popu, width,color='r', label='Bar graph1')
plt.bar(a, pop2, width,color='y', label='Bar graph2')

plt.title("Programming languages popularity graph")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")

plt.legend()
plt.show()
```

### creating horizontal bar graphs

```python
plt.barh(langs, popu, width,color='r', label='Bar graph1')
plt.barh(a, pop2, width,color='y', label='Bar graph2')

plt.title("Programming languages popularity graph")
plt.xlabel("Popularity")
plt.ylabel("Programming Languages")

plt.legend()
plt.show()
```
