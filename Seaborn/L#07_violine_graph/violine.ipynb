### A violin plot is a hybrid of a box plot and a kernel density plot. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable. The width of each curve corresponds with the approximate frequency of data points in each region.

```python
# import libaraies
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load dataset
d_set=sns.load_dataset('tips')
```

```python
# how many customers paid bills on each day visualisation
sns.violinplot(x='day', y='total_bill', data=d_set)
```

```python
# making use of hue
# to change color make use of palette parameter
# to split data according to hue parameter make use of split=true

sns.violinplot(x='day', y='total_bill', data=d_set, hue='time', palette='Dark2', split=True)
```
