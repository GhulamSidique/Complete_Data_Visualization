### in this notebook we will be wroking on creating bar plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load dataset
d_set=sns.load_dataset("penguins")
d_set
```

```python
# a simple bar plot
sns.barplot(x=d_set.island, y=d_set.bill_depth_mm)
# following code works similar as above
# sns.barplot(x="island", y="bill_depth_mm", data=d_set) 

plt.title("Sample bar plot")
plt.show()
```

```python
# add hue parameter
# to change the order of values make use of the order parameter 
# to change the porder of male and female make use of the hue_order parameter 
# to change the size of a small line over the bar graph make use of the errorbar=('ci', any number) OR use n_boot=any small number, parameter
# to change the color make use of the palette='name'
# to increase or decrease the color stauration make use of the stauration=any number
# to change the color of small line over bar graph make use of the errcolor='name' parameter
# to increase or decrease the width of the line over bar make use of the errwidth=any number parameter
# to add a cap over the bar plot along with its size make use of the capsize=any number


# to add grid
sns.set(style="darkgrid")

sns.barplot(x="island", y="bill_depth_mm", data=d_set, hue='sex', order=['Dream', 'Torgersen', 'Biscoe'], hue_order=['Female', 'Male'], errorbar=('ci', 100), palette='icefire', saturation=15, errcolor='b', errwidth=7, capsize=0.1) # ci ranges form 0 to 100

plt.title("Sample bar plot")
plt.show()
```

```python
# to create a horizontal bar graph we need to have bot x and y values as an integer not the string, and make use of the orient='horizontal or h' parameter

sns.barplot(x="bill_length_mm", y="bill_depth_mm", data=d_set, orient='h') # ci ranges form 0 to 100

plt.title("Sample bar plot")
plt.show()
```
