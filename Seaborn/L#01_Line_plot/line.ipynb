### Line plot using seaborn
### seaborn works on dataset and for this we need to have dataframe using pandas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# creating sample arrays
x=[1,2,3,4,5,6,7]
y=[2,3,4,1,5,6,8]

# plot aa simple line graph
sns.lineplot(x=x, y=y)

plt.show()
```

### Result:
#### In the above graph we did not make use of the dataframe and in the following graph we will be working with the dataframe and see the changes

```python
# same data but now with a dataframe
# creating sample arrays
x=[1,2,3,4,5,6,7]
y=[2,3,4,1,5,6,8]

# creating a dataframe
d_frame=pd.DataFrame({"x": x, "y": y})
# here x and  will be displayed as x and y labels automatically

# plot aa simple line graph
sns.lineplot(x="x", y="y", data=d_frame)

plt.show()
```

### working with a large dataset 

```python
d_set=sns.load_dataset("penguins").head(50)
d_set
```

```python
# now making comparisions between the columns from above dataset
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=d_set)
plt.show()
```

```python
# making use of the hue parameter that will plot graph in relation to the provideed column name as follows
# if we apply hue, legend will also automatically be applied
# to increase or decrease the size of the graph make use of the size parameter

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=d_set, hue='sex', size=10)
plt.show()
```

```python
# working with more parameters
# to change the style of the line graph make use of the style parameter as follows
# to change the color of graph make use of the palette parameter and write color name inside it. If you want to know the names of the color just write any A(or any other wrong name) inside palatte and a list of color names will be displayed  
# add marker to the graph make use of the markers=[], list is used because there are more than one values in the sex column 
# to remove thhe dotted line from the graph make use of the dashes=False parameter
# to remove legend make use of the legend=False parameter

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=d_set, hue='sex', style='sex', palette="Accent", markers=['*', 'o'], dashes=False, legend=False)

# to add title, grid make use of the matplotlib as follows
plt.title("Sample line graph")
plt.grid()

plt.show()
```
