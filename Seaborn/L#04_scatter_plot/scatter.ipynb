```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# load the data
data=sns.load_dataset("penguins")
data.head()
```

```python
# plot the scatter
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data, hue='sex')

plt.show()
```

```python
# to change the style of scatter data make use of the style='col name same as hue'
# to make variations in the sizes of the scatter data make use of the size='col name' parameter
# to change the size of the scatter make use of the sizes=(10, 10), the first number shows the x axis and other y axis 
# to change the color of scatter data make use of the palette="color name"


sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data, hue='sex', style='sex', size='sex', sizes=(20, 40), palette="gist_ncar")

plt.show()
```
